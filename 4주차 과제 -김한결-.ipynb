#### <span style="color:black">5. 주어진 6개의 숫자의 평균을 파이썬을 이용해서 구하시오</span>

#### <span style="color:black">6. 다음 주어진 4개의 숫자의 산술평균, 기하평균, 조화평균을 각각 파이썬을 이용해서 구하시오</span>
    4, 10, 20, 15

```python
산술평균 = (4+10+20+15)/4
기하평균 = (4*10*20*15)**(1/4)
조화평균 = 4/(1/4+1/10+1/20+1/15)
print("산술평균 :", 산술평균, "기하평균 :", 기하평균, "조화평균 :", 조화평균)
```
