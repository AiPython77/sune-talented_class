### 1. 주어진 리스트에서 최대값과 최소값을 동시에 찾는 알고리즘을 의사코드를 이용해서 작성하시오.

######주어진 리스트 [17, 92, 33, 58, 7, 33, 42]

    1. 가장 큰 수를 max라 하고 가장 작은 수를 min, 수 번째를 O, 리스트 항복 수를 n이라 하고 max, min에는 리스트의 첫 번째 수, O에 2, n에 주어진 리스트의 항목 수를 배정한다
    2. 주어진 리스트에서 O번째의 값이
      2-1. max보다 크다면
        2-1-1. max을 리스트의 O번째 값으로 설정한다.
      2-2. max보다 크지 않다면
        2-2-1. min보다 작다면
          2-2-1-1. min을 리스트의 O번째 값으로 설정한다.
        2-2-2. min보다 작지 않다면
         2-2-1-2. 통과한다.
    3. O에 1을 더한다.
    4. 위 과정을 O가 n-1이 같지 않는 동안 반복한다.
    5. max, min을 반환한다.

### 2. 위의 작성한 알고리즘의 계산복잡도는 무엇인가?
    O(n)입니다.

### 3. 작성한 알고리즘을 이용하여 리스트 v를 이용해서 최대값과 최소값을 동시에 찾는 파이썬 프로그램을 작성하시오.

```python
v = [17, 92, 18, 33, 58, 7, 33, 42]

max = v[0]
min = v[0]
O = 1
n = len(v)

while not O == n:
  if v[O] > max:
    max = v[O]
  else:
    if v[O] < min:
      min = v[O]
    else:
      pass
  O += 1
print("가장 큰 수는", max, "가장 작은 수는", min, "입니다.")
```

### 4. 다음 주어진 5개의 숫자의 평균, 표준편차 , 분산을 각각 파이썬을 이용해서 구하시오.

    주어진 숫자 [21, 37, 54, 48, 33]

```python
리스트 = [21, 37, 54, 48, 33]
횟수 = 0
분산 = 0

#평균 구하기
평균 = (21+37+54+48+33)/2

#분산 구하기
while not 횟수 == len(리스트):
  분산 += (평균-리스트[횟수])**2
  횟수 += 1
분산 = 분산/len(리스트)

#표준편차 구하기
표준편차 = 분산**(1/2)

print("평균은 " + str(평균) + "이고 표준편차는 " + str(표준편차) + ", 분산은 " + str(분산) + "입니다.")
```

### 5. 다음 주어진 숫자들의 최빈값을 파이썬을 이용해서 구하시오.

    주어진 숫자 [10, 8, 5, 4, 3, 6, 8, 7, 8, 6, 7, 9, 6, 8]

```python
리스트 = [10, 8, 5, 4, 3, 6, 8, 7, 8, 6, 7, 9, 6, 8]
최빈값 = '없다'
최대반복횟수 = 0

for 횟수 in list(set(리스트)):
  if 리스트.count(횟수) > 최대반복횟수:
    최빈값 = 횟수
    최대반복횟수 = 리스트.count(횟수)

print('최빈값: ' + str(최빈값)+ ', 반복횟수: ' + str(최대반복횟수))
```

### 6. 다음 주어진 숫자들의 중앙값을 파이썬을 이용해서 구하시오.

    주어진 숫자 [19, 82, 94, 28, 76, 59, 48, 31, 9]

```python
리스트 = [19, 82, 94, 28, 76, 59, 48, 31, 9]
리스트_길이 = len(리스트)
중앙값 = 0

리스트.sort()
if 리스트_길이%2 == 1:
  중앙값 =  리스트[int((리스트_길이-1)/2)]
else:
  중앙값 = (리스트[int(리스트_길이/2)-1]+리스트[int(리스트_길이/2)])/2
print("중앙값 :", 중앙값)
```
