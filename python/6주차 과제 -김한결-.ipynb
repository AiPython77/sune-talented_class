### 1. 1부터 n까지의 합 구하기를 재귀호출을 이용하여 파이썬 프로그램을 작성하시오.

    아래의 내용은 1부터 10까지를 계산기로 더한 내용입니다. 결과가 55가 나오는 것을 볼 수 있습니다. 아래의 내용을 파이썬 프로그램을 통해 재귀호출 함수인 합구하기 함수를 만들어 구해보겠습니다.

계산기 (소형).jpg

```python
def 합구하기(n):
  if n <= 1:
    return 1
  else:
    return n+합구하기(n-1)

###### 예) 1에서 10까지의 합 구하기(1에서 10까지의 합은 55입니다.)
print(합구하기(10))
```

### 3. 다음 주어진 숫자들의 범위를 파이썬을 이용해서 구하시오.

    주어진 숫자 [95, 88, 74, 11, 66, 45, 16, 7, 25, 37]

```python
주어진수 = [95, 88, 74, 11, 66, 45, 16, 7, 25, 37]
주어진수.sort()
가장큼 = 주어진수[len(주어진수)-1]
가장작음 = 주어진수[0]
범위 = 가장큼 - 가장작음

print("주어진 수들의 범위는 " + str(범위) + "입니다.")
```

### 4. 다음 주어진 숫자에서 60이상 80미만을 만족하는 숫자 빈도수를 파이썬을 이용해서 구하시오

    주어진 숫자 [26, 85, 45, 89, 74, 68, 61, 39, 58]

```python
주어진수 = [26, 85, 45, 89, 74, 68, 61, 39, 58]
빈도수 = 0
for 횟수 in range(len(주어진수)):
  if 주어진수[횟수] >= 60 and 주어진수[횟수] < 80:
    빈도수 += 1

print("주어진 숫자 중 60이상 80미만을 만족하는 숫자 빈도수는 " + str(빈도수) + "입니다.")
```

### 5. 다음 주어진 숫자에서 제 3사분위수를 파이썬을 이용해서 구하시오.

    주어진 숫자 [68, 45, 39, 26, 16, 77, 91, 61, 78, 88]

```python
def 백분위수(주어진수, 퍼센트):
  주어진수.sort()
  이전주어진수위치 = len(주어진수)*퍼센트/100
  주어진수위치 = round(이전주어진수위치)
  if 주어진수위치 == 이전주어진수위치:
    답 = (주어진수[주어진수위치-1]+주어진수[주어진수위치])/2
    return 답
  else:
    답 = 주어진수[주어진수위치-1]
    return 답

주어진수 = [68, 45, 39, 26, 16, 77, 91, 61, 78, 88]
print("주어진 숫자들의 제 3사분위수는" + str(백분위수(주어진수, 75)) + "입니다.")
```
